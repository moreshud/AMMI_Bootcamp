```python
#"title","region_2","taster_name","designation",
col_drop = ["index", "id"]
wine = wine.drop(col_drop, axis=1)
test = test.drop(col_drop, axis=1)
```

```python
wine.variety.iloc[86403] = "PINOT NOIR"
```

```python
wine["taster_twitter_handle"] = wine["taster_twitter_handle"].str.replace('@','')
test["taster_twitter_handle"] = test["taster_twitter_handle"].str.replace('@','')
```

```python
#country and province with null entries with price
coun_prov = wine[wine["country"].isnull()][wine["province"].isnull()]
coun_prov1 = test[test["country"].isnull()][test["province"].isnull()]
```

```python
coun_prov_price = coun_prov[coun_prov["price"].isnull()]
coun_prov_price1 = coun_prov1[coun_prov1["price"].isnull()]
```

```python
wine.drop(coun_prov_price.index, inplace=True)
test.drop(coun_prov_price1.index, inplace=True)
```

```python
wine["price"] = wine["price"].fillna(wine["price"].mean())
```

```python
import xgboost as xgb
from sklearn import model_selection, preprocessing
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
def clean_data(data, test_data):
    clean_df = data.copy()
    clean_df_test = test_data.copy()

    for f in clean_df.columns:
        if clean_df[f].dtype == 'object':
            label = preprocessing.LabelEncoder()
            label.fit(list(pd.concat([clean_df[f], clean_df_test[f]]).values))
            clean_df[f] = label.transform(list(clean_df[f].values))
            clean_df_test[f] = label.transform(list(clean_df_test[f].values))
    return (clean_df, clean_df_test)

wine_train, wine_test = clean_data(wine, test)
```

```python
wine_test = wine_test.drop(columns="price", axis =1)
```

```python
wine_test.head()
```

```python
wine_train.head()
```

```python
from sklearn.model_selection import train_test_split

X = wine_train.drop('price',axis=1)
y = wine_train['price']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
#Fitting XGB regressor 
model = xgb.XGBRegressor(n_estimators=10000, max_depth=10)
model.fit(X_train,y_train)
```

```python
from sklearn import metrics
predictions = model.predict(X_test)
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
output = model.predict(wine_test)
final_df = pd.DataFrame()
final_df["price"] = output
final_df.index.name = "id"
#final_df = final_df.set_index('id')
final_df.head()
```
